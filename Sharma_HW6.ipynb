## 3. SVD

## 3.1

```python
from scipy.linalg import svd as SVD
from skimage import io
import matplotlib.pyplot as plt
```

```python
img = io.imread('sjsu.png', as_gray = True)
```

```python
plt.imshow(img,'gray')
```

## 3.2 img array and size of array and calculate elements we should save for this image 

```python
print(img)
```

```python
img.shape
```

## # Number of elements to store = 1283 * 1200 = 1539600

## 3.3 SVD to find S, U, V

```python
U, S, V = SVD(img)
U.shape, S.shape, V.shape
```

```python
print(U)
```

```python
print(V)
```

```python
print(S)
```

```python
sigma = np.zeros((1283,1200))
for i in range(1200):
    sigma[i,i] = S[i]
print(sigma)
```

```python
AR = U@sigma@V
```

```python
print(np.round(AR))
```

```python
print(img)
```
