```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
```

```python
import warnings
warnings.filterwarnings("ignore")
```

### 1.a. Find “Concrete.csv” file and read it as data frame

```python
df = pd.read_csv('Concrete.csv')
df.head()
```

```python
df.info()
```

### 1.b. Choose “cement” as x (explanatory variable) and “csMPa” as y (response variable)

```python
#input
X = df[['cement']] #explanatory variable

#output
y = df[['csMPa']]  #response variable
```

```python
#model
slr = LinearRegression()
slr.fit(X,y)
```

### 1.c. Do a linear regression and find intercept and slope

```python
#B1*

slr.coef_

print("B1* : ", slr.coef_)

#B0*

slr.intercept_

print("B0* : ", slr.intercept_)
```

### 1.d. Draw scatter plot for all data and draw a line (scatter plot) you obtained in 1.c on the same graph

```python
plt.scatter(X,y)
plt.scatter(X,slr.predict(X), c='r')
plt.xlabel('cement')
plt.ylabel('csMPa')
plt.title('linear regression of cement vs csMPa')
```

### 1.e. Calculate R2 score

```python
R2_SLR = r2_score(y,slr.predict(X))
print("R2 : ",R2_SLR )
```

### 2.a. Find “Concrete.csv” file and read it as data frame

```python
df = pd.read_csv('Concrete.csv')
df.head()
```

### 2.b. Choose “cement” as x (explanatory variable) and “csMPa” as y (response variable). 

```python
#input
X = df[['cement']] #explanatory variable

#output
y = df[['csMPa']]  #response variable
```

### 2.c. Do a polynomial regression (with polynomial order = 3) and find intercept and slopes.

```python
polynomial_features = PolynomialFeatures(degree = 3, include_bias = False)
Xp = polynomial_features.fit_transform(X)
```

```python
pr = LinearRegression()
pr.fit(Xp, y)
```

```python
#B1*

pr.coef_

print("B1* : ", pr.coef_)

#B0*

pr.intercept_

print("B0* : ", pr.intercept_)
```

### 2.d. Calculate R2 score.

```python
R2_SLR = r2_score(y, pr.predict(Xp))
print("R2 : ",R2_SLR )
```

```python
plt.scatter(X, y)
plt.scatter(X, pr.predict(Xp), c='r')
plt.xlabel('Cement')
plt.ylabel('csMPa')
plt.title('Polynomial Regression of Cement vs csMPa')
```
